```python
gold_df = spark.table("insurance_project.gold_insurance")

display(gold_df)
```

```python
print("Total Records:", gold_df.count())
```

```python
train_df, test_df = gold_df.randomSplit(
    [0.8, 0.2],
    seed=42
)

print("Train Count:", train_df.count())
print("Test Count:", test_df.count())
```

```python
display(train_df)
```

```python
total = train_df.count() + test_df.count()

print("Combined Count:", total)
```

```python
train_df.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("insurance_project.train_insurance")
```

```python
test_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("insurance_project.test_insurance")
```
