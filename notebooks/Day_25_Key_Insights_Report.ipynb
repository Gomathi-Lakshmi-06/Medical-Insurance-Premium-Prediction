# Medical Insurance Premium Prediction
## Key Insights Report

## 1. Project Overview

The objective of this project is to predict medical insurance premiums using Machine Learning.

The dataset contains customer information such as age, gender, BMI, number of children, smoking status, and region.

Random Forest Regression was used to build the prediction model.

## 2. Smoker vs Non-Smoker Analysis

Smoking status has the strongest impact on insurance premiums.

From the dashboard and prediction results, smokers consistently show much higher insurance costs compared to non-smokers.

The machine learning model identified smoker status as one of the most important features affecting premium prediction.

Key Finding:
- Smokers pay significantly higher premiums.
- Smoking is the strongest cost-driving factor.

## 3. BMI Impact Analysis

BMI (Body Mass Index) has a noticeable effect on insurance premiums.

As BMI increases, premiums generally increase because higher BMI is associated with greater health risks.

Customers with obesity tend to have higher medical expenses and insurance charges.

Key Finding:
- Higher BMI leads to higher premiums.
- BMI is one of the important predictive factors.

## 4. Age Impact Analysis

Age has a positive relationship with insurance premiums.

As customers grow older, medical risks increase, resulting in higher insurance costs.

The dashboard shows that premium values generally increase with age.

Key Finding:
- Older individuals have higher insurance premiums.
- Age is an important predictor in the model.

## 5. Regional Analysis

Premium amounts vary slightly across regions.

However, regional differences are smaller compared to smoking status and age.

The region-based chart indicates moderate variations in average insurance costs.

Key Finding:
- Regional impact exists.
- Region is less influential than smoking and age.

## 6. Impact of Number of Children

The number of children has a relatively small effect on insurance premiums.

While some increase in premium can occur with more dependents, the impact is much lower compared to smoking status, age, and BMI.

Key Finding:
- Children count has limited influence.
- Other factors contribute more strongly to premium changes.

## 7. Machine Learning Model Findings

Random Forest Regression was used for premium prediction.

The model successfully learned patterns from customer data and generated realistic premium estimates.

Feature importance analysis indicated that smoking status, age, and BMI were the most influential factors.

Key Finding:
- Random Forest Regression performs well for premium prediction.
- The model captures important customer risk factors.

## 8. Conclusion

The Medical Insurance Premium Prediction project successfully identified the major factors affecting insurance charges.

Smoking status was found to be the strongest contributor to higher premiums.

Age and BMI also showed significant influence on insurance costs.

The Random Forest Regression model successfully predicted premiums and provided meaningful insights for insurance pricing.

The project demonstrates how Machine Learning can assist in risk assessment and premium estimation.
